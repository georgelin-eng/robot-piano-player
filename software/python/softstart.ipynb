```python
import numpy as np
import matplotlib.pyplot as plt

def smoothstep(t_ramp, t):
    x = np.array(t)
    x = np.clip(x, None, t_ramp)
    u = 3*(x/t_ramp)**2 - 2*(x/t_ramp)**3

    return u

num_setpoints = 10
setpoints = np.random.randint(low=0, high=20, size=num_setpoints)

t_ramp = 0.3
t_per_setpoint = 0.5

t = np.linspace(0, t_per_setpoint, 1000)

a = 2*np.pi / t_ramp

curr_setpoint = []
for setpoint in setpoints:

    yd = setpoint

u = smoothstep(t_ramp, t)

plt.grid()
plt.plot(t, u)
plt.show()
```
