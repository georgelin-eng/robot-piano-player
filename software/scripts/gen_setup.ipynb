```python
# import pandas as pd
from pandas import read_csv

df = read_csv("IO.csv", delimiter="\t")

f = open("setup.txt", "w")

for i, row in df.iterrows():
    pin_name  = row["Pin"]
    direction = row["Direction"]
    Variable_name = row["Variable_name"]

    output_txt = f"pinMode({pin_name}, {direction}); // {Variable_name}\n"   
    f.write(output_txt)

```
